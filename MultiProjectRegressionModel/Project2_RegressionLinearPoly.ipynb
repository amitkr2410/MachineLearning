### Build a linear and polynomial regression model using sci-kit learn library
This is regression model to predict the expected salary in a company for a given position/level


```python
#Load pythonn library:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load the data set
dataset = pd.read_csv('Project2_RankInCompany_Vs_Salary.csv') 
print(dataset)

#Create training data set for Linear and polynomial regression
X = dataset.iloc[:,1:2].values
y = dataset.iloc[:,-1].values
print(X)
print(y)
```

```python
#Linear regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
print(X)

y_predict = lin_reg.predict(X)
print(y_predict)

#Polynomial regression
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X)
print(X_poly)

lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, y)
y_predict_poly = lin_reg2.predict(X_poly)
print(y_predict_poly)
```

```python
plt.scatter(X,y, color='red')
plt.plot(X, y_predict, color='blue')
plt.plot(X, y_predict_poly, color='black')
plt.xlabel('Rank in company')
plt.ylabel('Salary')
x1 = poly_reg.fit_transform([[6]])
print(x1)
x0 = np.array( [[ 1.0, 6.0,  36.0]] )
print(x0)
y0 = lin_reg2.predict(x0 )
print(y0)
```
