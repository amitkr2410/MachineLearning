### We create python script to download resumes from livecareer.com
### for example :
###    https://www.livecareer.com/resume-search/search?jt=hr&bg=85&eg=100&comp=&mod=&pg=1
### using webscrapping tools. The extracted html code is converted using pdfkit and saved into pdf files
### The extracted html code is also saved into a csv file

```python
# Import libraries
# BeautifulSoup for webscrapping with html elements
# import requests module to make the HTTP request which returns a Response Object with all the response data
from bs4 import BeautifulSoup
import bs4
import requests
from selenium import webdriver
import os
import pdfkit

import pandas as pd
import numpy as np
```

```python
#driver = webdriver.Chrome("chromedriver.exe")

# LIST OF UNIQUE CATEGORIES THAT WAS NEEDED FOR THE DATASET.
# ADD OR SUBTRACT LABELS AS NEEDED
job_list = ['HR', 'Data-Scientist', 'Machine-Learning-Engineer',  'TEACHER', 'ADVOCATE',
 'Dental-Hygienist', 'Doctor', 'HEALTHCARE', 'FITNESS', 'AGRICULTURE', 'SALES',
 'CONSULTANT', 'DIGITAL-MEDIA', 'AUTOMOBILE', 'CHEF', 'FINANCE', 'APPAREL',
 'ENGINEERING', 'ACCOUNTANT', 'CONSTRUCTION', 'PUBLIC-RELATIONS', 'BANKING',
 'ARTS', 'AVIATION']
```

```python
# CREATING A DATAFRAME TO STORE THE LINKS OF EACH INDIVIDUAL RESUME EXAMPLE

df_resume = pd.DataFrame()
category = []
link = []
```

```python
driver = webdriver.Chrome()

for job in job_list:
    JOB = job.lower()
    for i in range(1,2):   # INCREASE THE RANGE TO GET MORE RESUME DATA
        PAGE = str(i)
        url= "https://www.livecareer.com/resume-search/search?jt=" + JOB + "&bg=85&eg=100&comp=&mod=&pg=" + PAGE
        print(url,"\n")
        driver.get(url)
        aTagsInLi = driver.find_elements_by_css_selector('body a')
        #print(aTagsInLi)
        
        Iter=0
        for a in aTagsInLi:
            #print("AmitLine=", Iter)
            #print(a)
            #print(a.get_attribute('href'))
            #print(a.get_attribute('rel'))
            #print(a.get_attribute("noopener noreferrer"))
            if a.get_attribute('rel') == "noopener noreferrer" and a.get_attribute('class')!="fr-link":
                if a.get_attribute('class')!="byr-link":
                    category.append(JOB)
                    link.append(a.get_attribute('href'))
                    print(a.get_attribute('href'))
            Iter=Iter+1
        
```

```python
# STORES THE CATEGORY AND LINK TO THE RESUME WEBPAGE

df_resume["Category"] = category
df_resume["link"] = link
```

```python
# HASHES THE LINK AND CREATES AN UNIQUE ID FOR THE ROWS

import hashlib
def id(x):
    #ans=hashlib.md5(x.encode('utf-8')).hexdigest()
    mystring = x.split('?')[0]
    Ans = mystring.split('-')[-1]
    ans = int(Ans, 16) 
    ans = str(ans)[:6]
    #ans=hashlib.md5(ans.encode('utf-8')).hexdigest()
    print(x, Ans, ans)
    return ans
    #return ans


df_resume["id"] = df_resume["link"].apply(id)
```

```python
# LOOP TO COLLECT THE RESUME DATA PRESENT IN THE LINKS AS COLECTED ABOVE
# STORED IN HTML AS WELL AS STRING FORMAT
import os
import pdfkit
config = pdfkit.configuration(wkhtmltopdf="/usr/local/bin/wkhtmltopdf")
parent_dir ="data/"

for i in range(df_resume.shape[0]):
#for i in range(1):
    url = df_resume.link[i]
    driver.get(url)
    #print(url)
    
#     time.sleep(0.5)                  #ADDED DELAY, CAN BE REMOVED
    x = driver.page_source
    #x = x.replace(">","> ")
    scrapped = bs4.BeautifulSoup(x, 'html.parser')
    #print(scrapped)
    div = scrapped.find(id="server-resume-template")
    #div = scrapped.find("div", {"id": "document"})
    #print(div)
  
    df_resume.Raw_htmlresume[i] = str(div)
#    try:
#        df_resume.ResumeTitle[i] = a.text
#    except:
#         ADD EXCEPTION IF REQUIRED
#        pass
    path = os.path.join(parent_dir, df_resume.Category[i].upper() )
    os.makedirs(path, exist_ok=True)
    filename = path + "/" + str(df_resume.id[i]) + ".pdf"
    pdfkit.from_string(str(div), filename, configuration = config) 
    
#How to use pdfkit and create pdf file
#import pdfkit
#config = pdfkit.configuration(wkhtmltopdf="/usr/local/bin/wkhtmltopdf")
#pdfkit.from_url('http://google.com', 'out.pdf',configuration = config)
#pdfkit.from_file('test.html', 'out2.pdf',configuration = config)
#pdfkit.from_string('Hello!', 'out3.pdf',configuration = config)
   
```
